# Gradient Descent

```python
#importing required depedencies
import numpy as np                  #for numerical calculation
import matplotlib.pyplot as plt     #for data visualization
```

```python
#Generate sample data using numpy random function
m = 100   # Sample size
x = 2*np.random.rand(m,1)           #generate sample data for x                        
y = 4 + 3*x + np.random.rand(m, 1)  #generate sample data for y
x_b = np.c_[np.ones((m,1)), x]      #add bias X0 = 1
x_new = np.array([[0],[2]])         # making prediction at x_new value
x_new_b = np.c_[np.ones((2,1)), x_new] # adding bias to x_new value
```

## Impelement gradient descent

```python
theta = np.random.rand(2,1)         #random initialization of theta
n_iteration = 10                    #Define number of iterations
```

## learning rate eta = 0.1

```python
for i in range(n_iteration):                           #iterate over range of number of iteration
    eta = 0.1                                          #learning rate
    gradient = 2/m * (x_b.T).dot(x_b.dot(theta)- y)    #calculating batch gradient
    theta = theta - eta * gradient                     #calculating next theta
    y_predict = x_new_b.dot(theta)                     #predict value 
    plt.plot(x_new, y_predict, 'b-')                   #plot predicted value
    plt.scatter(x, y)                                  #plot scatter plot of give input x, y
    plt.xlabel('x-axis')                               #define axis name
    plt.ylabel('y-axis')                               #define axis name
    plt.legend([f"Learning rate: {eta}"])              #define plot legend
    plt.title(f'Gradinet descent for learning rate eta = {eta}')  #define plot title

theta
```

## learning rate eta = 0.5

```python
"""
Unable to attain global minima, because of high learning rate
"""
for i in range(n_iteration):                                #iterate over range of number of iteration
    eta = 0.5                                               #learning rate            
    gradient = 2/m * (x_b.T).dot(x_b.dot(theta)- y)         #calculating batch gradient
    theta = theta - eta * gradient                          #calculating next theta
    y_predict = x_new_b.dot(theta)                          #predict value 
    plt.plot(x_new, y_predict, 'b-')                        #plot predicted value
    plt.scatter(x, y)                                       #plot scatter plot of give input x, y
    plt.xlabel('x-axis')                                    #define axis name
    plt.ylabel('y-axis')                                    #define axis name
    plt.legend([f"Learning rate: {eta}"])                   #define plot legend
    plt.title(f'Gradinet descent for learning rate eta = {eta}')  #define plot title

theta
```

## learning rate eta = 0.01
