## Stochastic gradient descent

```python
#importing required depedencies
import numpy as np                  #for numerical calculation
import matplotlib.pyplot as plt     #for data visualization

# setting matplotlib plotting parameters
plt.rcdefaults()
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['font.size'] = 12
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 10
plt.rc('font', family='serif')
```

```python
#Generate sample data using numpy random function
sample_size = 100                             # Sample size
x = 2*np.random.rand(sample_size,1)           #generate sample data for x                        
y = 4 + 3*x + np.random.rand(sample_size, 1)  #generate sample data for y
theta = np.random.rand(2,1)                   #random initialization of theta
x_b = np.c_[np.ones((sample_size,1)), x]                #add bias X0 = 1
x_new = np.array([[0],[2]])                   # making prediction at x_new value
x_new_b = np.c_[np.ones((2,1)), x_new]        # adding bias to x_new value
```

```python
#created a function called learning_schedule to gradually decrease
#learning rate
def learning_schedule(t):
    return t0 / (t + t1)
```

```python
n_epochs = 10                                                   #define number of epochs
t0, t1 = 5, 50                                                  #define learning schedule parameter

for epoch in range(n_epochs):                                   # iterate over range of number of epochs
    for j in range(sample_size):                                          # iterate over range of number of sample 'm'
        number_index = np.random.randint(sample_size)                     # selecting random index
        xi = x_b[number_index:number_index+1]                   # select data at aprticular index for x
        yi = y[number_index:number_index+1]                     # select data at aprticular index for y
        gradient = 2  * (xi.T).dot(xi.dot(theta) - yi)          # calculate gradient
        eta = learning_schedule(epoch * (sample_size + j))                # Gradually decreasing the learning rate
        theta = theta - eta * gradient                          # Calculate next theta
        if j == 1:                                              #for first random instance
            y_predict = x_new_b.dot(theta)                      #calculate prediction
            plt.plot(x_new, y_predict, label = f"epoch:{epoch +1 }")    #plot predicted data
            plt.scatter(x,y)                                    #generat scatter plot for input data
            plt.xlabel('x1')                                    #Define x axis name
            plt.ylabel('y')                                     #Define y axis name
            plt.title('Stochastic Gradient Descent')            # Define title of plot
            plt.legend()

    
print(f"Final epoch theta: {theta[0], theta[1]}")
```
