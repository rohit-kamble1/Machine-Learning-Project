```python
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt

os.chdir("E:\\pandas")
```

```python
df_tips=sns.load_dataset("tips")
df_tips.head()
```

```python
tc=df_tips.corr()
tc
```

```python
sns.heatmap(tc, annot=True)
```

```python
df_flight=sns.load_dataset("flights")
df_flight.head()
```

```python
fc= df_flight.corr()
fc
```

```python
fc= df_flight.pivot_table(index='month', columns="year", values='passengers')
fc
```

```python
sns.heatmap(fc, cmap="coolwarm")
```
