#### Import Necessary Libraries

```python
# Import dependencies  
import pandas as pd                 #for data manipulation
import numpy as np                  #for numerical operations
import matplotlib.pyplot as plt     #for data visualisation
import seaborn as sns               #for data visualisation
```

#### Sample dataframe

```python
#read csv file using pandas datafram
df = pd.read_csv('https://raw.githubusercontent.com/Baakchsu/LinearRegression/master/weight-height.csv')  
df.head()  #First five rows
```

```python
#Get dataframe information
df.info()
```

#### Let's understand each type of plot one by one

##### Scatter Plot
*  Scatter plot is a set of points that represnets the value obtained for two different variables plotted on a horizontal and vartical axes

> When to used "scatter plot"??
* Scatter plot are used to convey the relationship between two numerical variables.
* Scatter plots are somtimes called co-relation plots because they show how two variables are co-related.

##### Scatter plot for height and weight

```python
X = df[['Height']]  #define column height on x axis
y = df[['Weight']]  #define column weight on y axis 
```

```python
#Using matplotlib we can plot scatter plot
#size of each data point is 10 and color of data point is green.
#you can change the size of data points and color using hyperparameter s and color.

plt.scatter(X, y, s=10, color="green") 
plt.xlabel("Height")                            #define x-axis variable name
plt.ylabel("weight")                            #define y-axis variable name
plt.title("Scatter plot of Height & Weight")    #define title of plot
plt.legend(['Dataset'])                         #define legend of dataset   
plt.show()                                      #show the plot
```

##### Histogram
* Histogram is graph of that shows the distribution of data. Histogram are often used to illustrate the main feature data distribution. Here we can use histogram to show data distribution among different column.

> When to used?
* A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.

##### Histogram of height

```python
#plot histogram using matlplotlib.pyplot of height column.
#The hyperparamter color define color of bars, bins defines number of bars in graph with help of edgecolor we can 
#seprate each bar with whitespace 
plt.hist(df[['Height']],edgecolor='white',color="green",bins=20, label="Height")
plt.title("Histogram of Height")    #define plot title
plt.xlabel('Height')                #define name on x-axis
plt.ylabel('Frequency')             #Frequncy of each bar distribution
plt.legend()                        
plt.show()                          
```

##### Histogram of weight

```python
#plot histogram using matlplotlib.pyplot of weight column.
#The hyperparamter color define color of bars, bins defines number of bars in graph with help of edgecolor we can 
#seprate each bar with whitespace 
plt.hist(df[['Weight']],edgecolor='white',color="orange",bins=20, label="Weight")
plt.title("Histogram of Weight")      #define plot title
plt.xlabel('Weight')                  #define name on x-axis
plt.ylabel('Frequency')               #Frequncy of each bar distribution
plt.legend()
plt.show()
```

#### Let's understand each type of plot one by one

##### Scatter Plot
* Scatter plot is a set of points that represnets the value obtained for two different variables plotted on a horizontal and vartical axes

> When to used "scatter plot"??
* Scatter plot are used to convey the relationship between two numerical variables.
* Scatter plots are somtimes called co-relation plots because they show how two variables are co-related.

```python
 #set_style parameter to control the general tyoe of plots. We can use other style such as darkgrid, whitegrid, dark, white, ticks.
sns.set_style(style="darkgrid")
#plot scatterplot
sns.scatterplot(data = df, x = df["Height"], y = df["Weight"]) 
```

* Scatter plot contains hyperparamter called hue, which groups the variable that will produce points in different colors. It can be
categorical or numeric

```python
sns.scatterplot(data = df, x = df["Height"], y = df["Weight"], hue = 'Gender') 
```

* Assign the same variable to another hyperparamter called style to change markers of data.

```python
sns.scatterplot(data = df, x = df["Height"], y = df["Weight"], hue = 'Gender', style= 'Gender') 
```

* We can control the specific markers used to map the style variable by passing a Python list or dictionary of marker codes

```python
markers = {"Male": "X", "Female": "o"}
sns.scatterplot(data = df, x = df["Height"], y = df["Weight"], hue = 'Gender', style= 'Gender', markers = markers) 
```

##### Bar Plot

* Show point estimates and errors as rectangular bars.

> When to use bar plot?
* If we want to represent aggregate or statistical estimate for numerical variable with the height of each rectangle.

```python
data = sns.load_dataset('penguins')  #load data
data.tail(10)
pd.unique(data['species'])
```

```python
data.info()
```

```python
# assign x = 'island' and y = 'body_mass_g' to group by a categorical varaible and plot aggregated values.
sns.barplot(data = data, x = 'island', y = 'body_mass_g' )  
plt.title("Aggregated body mass with respect to island")
```

* We can add a second layer of grouping with hue hyperparameter

```python
sns.barplot(data = data, x = 'island', y = 'body_mass_g', hue = 'sex')  
plt.title("Aggregated body mass of male, female penguins with respect to island")
```

##### Heatmap
* Plot rectangular data as a color-encoded matrix.

> When to used heatmap?
* Heatmap are often used to show a generalised view of large amounts of data. Colors are easier to understand than raw numbers.

```python
data = sns.load_dataset("glue")    # load dataset
data.head(10)
```

```python
data.info()   #check data information
```

```python
#plot heatmap
data = data.pivot(index="Model", columns="Task", values="Score")
data.head()
```

```python
sns.heatmap(data)
```

* Use annot to represent the cell values with text

```python
sns.heatmap(data, annot = True)
```

* Control the annotations with a formatting string

```python
sns.heatmap(data, annot=True, fmt=".1f")
```

* Add lines between cells

```python
sns.heatmap(data, annot=True, linewidth=.5)
```

* Choose a different colormap by name

```python
sns.heatmap(data, cmap="crest", linewidth=.5)
```

##### Hist plot
* Plot univariate or bivariate histograms to show distributions of datasets.

> When to used hist plot?
* A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.


```python
sns.histplot(data = df, x = 'Height', bins = 20)  #histogram for height
```

* Flip the plot by assigning the data variable to the y axis

```python
sns.histplot(data = df, y = 'Height', bins = 20) 
```

* Add kernel density estimator.

```python
sns.histplot(data = df, x = 'Height', bins = 20, kde = True) 
```

##### Box Plot
* Draw a box plot to show distributions with respect to categories.

> When to use box plot?
* Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data’s **symmetry, skew, variance, and outliers**.

```python
sns.boxplot(data = df, x = 'Height', hue = 'Gender')
```

##### Regplot

* Plot data and a fit a linear regression model

```python
sns.regplot(data = df, x = 'Height', y = 'Weight', line_kws=dict(color="r"))
```
