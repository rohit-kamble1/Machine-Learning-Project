```python
df.dropna(inplace=True)
```

```python
x=df["Sentence"]
y=df["Sentiment"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
```

```python
model_clf=Pipeline([("tfid", TfidfVectorizer()),("clf", DecisionTreeClassifier())])
```

```python
model_clf.fit(x_train, y_train)
```

```python
y_pred=model_clf.predict(y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
model_clf.predict(["This is good movie"])
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
df["score"]=df["Sentiment"].apply(lambda Sentiment: sid.polarity_scores(Sentiment))
df.head(3)
```

```python
df["compound"]=df["score"].apply(lambda d:d["compound"])
df.head(3)
```

```python
#lambda <args> : <return Value> if <condition > ( <return value > if <condition> else <return value>)
#converter = lambda x : x*2 if x < 11 else (x*3 if x < 22 else x)
df["com_score"]=df["compound"].apply(lambda c: 'positive' if c>0 else ('negative' if c<0 else 'neutral'))
df.head(3)
```

```python
df['score'][1]
```

```python
from sklearn.metrics import classification_report
print(classification_report(df['Sentiment'],df['com_score']))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(df['Sentiment'],df['com_score']))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(df['Sentiment'],df['com_score']))
```
