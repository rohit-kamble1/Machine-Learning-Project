```python
from tensorflow import keras
import numpy as np
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
#Data preprocessing
train_datagen=ImageDataGenerator(
                                rescale=1./255,
                                shear_range=0.2,
                                zoom_range=0.2,
                                horizontal_flip=True
)


training_set=train_datagen.flow_from_directory(
                            '../input/chest-xray-pneumonia/chest_xray/train',
                            target_size=(64,64),
                            batch_size=32,
                            class_mode='binary'
)
```

```python
#preprocessing the test data
test_datagen=ImageDataGenerator(rescale=1./255)

test_set=test_datagen.flow_from_directory(
                        "../input/chest-xray-pneumonia/chest_xray/test",
                        target_size=(64,64),
                        batch_size=32,
                        class_mode='binary'
)
```

```python
#Building CNN
##Initialize the CNN

model=tf.keras.models.Sequential()
```

```python
#Adding comvolutional layer
model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64,64,3]))
#adding poiling layer
model.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
model.add(tf.keras.layers.Dropout(0.5))

#Adding one more convolutional layer with pooling layer
model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
model.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
model.add(tf.keras.layers.Dropout(0.5))
```

```python
#Flattening
model.add(tf.keras.layers.Flatten())
```

```python
#Full connection
model.add(tf.keras.layers.Dense(units=128,activation="relu"))
```

```python
#output layer
model.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

```python
#compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
```

```python
#Training the cnn model on training set and evaluating test set
model.fit(x=training_set, validation_data=test_set, epochs=25)
```
