```python
df.dropna(inplace=True)
```

```python
df.info()
```

**SENTIMENT ANALYSIS**

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
df["score"]=df["review"].apply(lambda review: sid.polarity_scores(review))
df.head(3)
```

```python
df["compound"]=df["score"].apply(lambda d:d["compound"])
df.head(3)
```

```python
df["com_score"]=df["compound"].apply(lambda c: 'positive' if c>=0 else 'negative')
df.head(3)
```

```python
from sklearn.metrics import classification_report
print(classification_report(df['sentiment'],df['com_score']))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(df['sentiment'],df['com_score']))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(df['sentiment'],df['com_score']))
```

**TEXT ANALYSIS**

```python
X=df["review"]
y=df["sentiment"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
```

```python
test_clf=Pipeline([("tfid",TfidfVectorizer()),("clf",LinearSVC())])
test_clf.fit(X_train,y_train)
```

```python
y_pred=test_clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
